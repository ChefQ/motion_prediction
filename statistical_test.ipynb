```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt
```

```python
betas = []
probabilities =  np.arange(0,1.1,0.01)
for i in probabilities:
    beta = np.sum( [ binom.pmf(x,15,i) for x in range(3,16)])
    betas.append(beta)

fig, ax = plt.subplots()
ax.set_title("Beta vs. Probability")
ax.set_xlabel("Probability")
ax.set_ylabel("Beta")
ax.plot(probabilities,betas)
```

```python
betas = []
Ks =  np.arange(3,16,1)
for k in Ks:
    beta = np.sum( [ binom.pmf(x,15,0.3) for x in range(k,16)])
    betas.append(beta)

fig, ax = plt.subplots()
ax.set_title("Beta vs. K")
ax.set_xlabel("K")
ax.set_ylabel("Beta")
ax.plot(Ks,betas)
```
